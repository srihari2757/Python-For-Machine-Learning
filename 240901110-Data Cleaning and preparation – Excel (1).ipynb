```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("C:\\Users\\DELL\\Downloads\\II EEE marks.xlsx")
print(df)
```

```python
import pandas as pd
df.columns = df.columns.str.strip().str.upper()
print("Columns found in your file:", df.columns.tolist())
name_col = [col for col in df.columns if 'NAME' in col]

if not name_col:
    print("Error: No column containing 'NAME' was found. Please check your Excel file headers.")
else:
    target_column = name_col[0]
    f = df[target_column].astype(str).str.strip()
    f.to_excel("C:\\Users\\DELL\\Downloads\\II EEE marks.xlsx")

    print("\nReplace value:")
    k = df.ffill() 
    print(k)

    print("\nUpdated list\n")
    k.to_excel("C:\\Users\\DELL\\Downloads\\II EEE marks.xlsx")
```

```python
print("\nDrop one particular column and its values:\n")
k.drop(['MATHS'],axis=1,inplace=True)
print(k)
print("\nUpdated list..\n")
k.to_excel("C:\\Users\\DELL\\Downloads\\II EEE marks.xlsx")
```
