```python
import datetime
cd=datetime.datetime.now()
print(cd)
```

```python
res=cd.strftime("%y")
print("\nTwo digit year:",res)
```

```python
res1=cd.strftime("%Y")
print("\nFour-digit year:",res1)
```

```python
res=cd.strftime("%m")
print("\nTwo digit month[01,12]:",res)
```

```python
res=cd.strftime("%b")
print("\nShort version of month",res)
```

```python
res=cd.strftime("%B")
print("\nFull name of the month:",res)
```

```python
res=cd.strftime("%j")
print("\nDays of the year:",res)
```

```python
res=cd.strftime("%D")
print("\nShortcut for %m/%d/%y:",res)
```

```python
res=cd.strftime("%d")
print("\nTwo digit day:",res)
```

```python
res=cd.strftime("%a")
print("\nShort version of the day",res)
```

```python
res=cd.strftime("%A")
print("\nFull name of the day:",res)
```

```python
res=cd.strftime("%H")
print("\nHour (24 hour clock):",res)
```

```python
res=cd.strftime("%I")
print("\nHour (12-hour clock)",res)
```

```python
res=cd.strftime("%M")
print("\nTwo digit minuite:",res)
```

```python
res=cd.strftime("%S")
print("\nSecond:",res)
```

```python
res=cd.strftime("%F")
print("\nShortcut for%Y-%m-%d:",res)
```

```python
res=cd.strftime("%f")
print("\nMicrosecond as an integer:",res)
```

```python
res=cd.strftime("%p")
print("\nLocal equivalent of AM or PM:",res)
```

```python
res=cd.strftime("%x")
print("Local appropriate formatted date:",res)
```

```python
res=cd.strftime("%X")
print("\nLocal appropriate time:",res)
```
