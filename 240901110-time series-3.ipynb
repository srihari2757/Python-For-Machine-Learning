```python
import pandas as pd
import numpy as np
from datetime import datetime
dates=[datetime(2011,1,2),
          datetime(2011,1,5),
          datetime(2011,1,7),
          datetime(2011,1,8),
          datetime(2011,1,10),
          datetime(2011,1,12)]
ts=pd.Series(np.random.random(6),index=dates)
print("Date Index\n",ts)
```

```python
print("\nDate index alone from the series:\n",ts.index)
```

```python
stamp=ts.index[2]
print("\nSecond index:\n",stamp)
```

```python
longer_ts=pd.Series(np.random.randn(1000),
                    index=pd.date_range("2000-01-01",periods=1000))
print("\nDate range upto 1000 days:\n",longer_ts)
                                        
```

```python
print("\nDate range-2001 year:\n",longer_ts["2001"])
```

```python
print("Date range=05 month 2001 year:\n",longer_ts["2001-05"])
```

```python
print("\n Date range-first 15 dates in 02 month 2001 year:\n",longer_ts["2000-02"].head(15))
```

```python
print("\nDate range-from 26-09-2001\n",longer_ts[datetime(2000,9,26):])
```

```python
print("\nDate   range   from    26-9-2001   to  25-9-2002:\n",longer_ts[datetime(2001,9,26):datetime(2002,9,25)])
```
